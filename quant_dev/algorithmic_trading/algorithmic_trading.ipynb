```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
first = '2020-01-01'
```

```python
tickers = [
    'MSFT',# equities: microsoft, tesla, nvidia, home depot
    'TSLA',
    'NVDA',
    'HD',
    'GC=F',# commodities: gold, copper, crude oil, natural gas
    'HG=F',
    'CL=F',
    'NG=F',    
]
```

```python
data = yf.download(tickers, start=first)
```

```python
data.info()
```

```python
file = 'data.csv'
data.to_csv(file)
data=pd.read_csv(file)
```

```python
data = pd.read_csv(file,header=[0,1],index_col=[0], parse_dates=[0])
data
```

```python
data.columns
```

```python
data.columns=pd.MultiIndex.from_tuples(data.columns)
data
```

```python
data.describe()
```

```python
close = data.loc[:,'Close'].copy()
close
```

```python
%matplotlib inline
plt.style.use('seaborn-v0_8')
```

```python
close.plot(figsize=(16,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
# norm_close = close.div(close.iloc[0]).mul(100)
norm_close = close.div(close.iloc[0])
norm_close
```

```python
norm_close.plot(figsize=(12,6), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,['mean', 'std']]
summary
```

```python
summary['mean'] = summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x='std',y='mean',figsize=(12,4),s=50,fontsize=12)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std'],summary.loc[i,'mean']),size=12)
plt.xlabel('Annual Risk(std)', fontsize=12)
plt.ylabel('Annual Return')
plt.title('Return vs Risk', fontsize=16)
plt.show()
```

## SMA: 10, 50, 200

```python
# sma_10 = close['NG=F'].rolling(window=10,min_periods=10).mean().dropna()
# sma_200 = close.rolling(window=200,min_periods=200).mean().dropna()
sma_50 = close['CL=F'].rolling(window=50,min_periods=50).mean().dropna()
sma_50.plot(figsize=(8,4), fontsize=12)
plt.title('Crude Oil Futures', fontsize=16)
plt.ylabel('Price ($)')
plt.show()
```
